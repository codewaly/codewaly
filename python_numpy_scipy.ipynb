##### $1.$  Écrire une fonction nextpower qui calcule la première puissance de $2$ supérieure ou égale à un nombre $n.$

```python
def nextpower(n):
    k=0
    nextpower=1
    while n > nextpower :
        k=k+1
        nextpower=2**k
    return k
```

```python
print(nextpower(5))
```
